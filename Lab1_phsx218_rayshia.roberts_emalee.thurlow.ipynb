$\alpha = \frac{\Delta L}{L*\Delta T}$

```python
import numpy as np

def rule3(errA,errB):
    errQ = np.sqrt(errA**2 + errB**2)
    return errQ

errT_i = 0.3 #C
errT_f = 0.1 #C

err_Delta_T = rule3(errT_i, errT_f)
print ("Error in change in Temperature =",err_Delta_T, "degrees C")
```

```python
def rule4(A,B,C,errA,errB,errC,n,m):
    errQ = np.sqrt(errA/A**2 + n*errB/B**2 + m*errC/C**2)
    return errQ

err_Delta_L = .01 #mm
err_Delta_T = .316 #C
errL = 1.0 #mm
L = 1050 #mm
Delta_L = 1.25 #mm
Delta_T = 69.1 #C
n = -1
m = -1

err_Alpha = rule4(Delta_L, L, Delta_T, err_Delta_L, errL, err_Delta_T, n, m)
print ("Error in alpha =",err_Alpha, "per degrees C")
```
